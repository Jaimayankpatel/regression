```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from tensorflow import math
from tensorflow.keras import datasets,layers,models
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.cifar10.load_data()
```

```python
y_test=y_test.reshape(-1,)
y_test[:10]
```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
flat_x_train=x_train.reshape(len(x_train),32*32*3)
flat_x_test=x_test.reshape(len(x_test),32*32*3)
flat_x_train
flat_x_test
```

```python
model = keras.Sequential([
    keras.layers.Dense(10,input_shape=(3072,),activation="sigmoid")])
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
model.fit(flat_x_train,y_train,epochs=5)
```

```python
y_predict=model.predict(flat_x_test)
y_predict[7089]
```

```python
y_predict=model.predict(flat_x_test)
y_predict[7089]
```

```python
plt.imshow(x_test[7089])
```

```python
np.argmax((y_predict[7089]))
```

```python
y_predict_all=[np.argmax(i) for i in y_predict]
y_predict_all
```

```python
abc=tf.math.confusion_matrix(labels=y_test,predictions=y_predict_all)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(abc,annot=True,fmt="d")
plt.title("my beautiful graph")
plt.xlabel("pridiction")
plt.ylabel("test")
```

```python
cnn = models.Sequential([
    #conv
    layers.Conv2D(filters=32, kernel_size=(3,3),activation="relu",input_shape=(32,32,3)),
    layers.MaxPooling2D(2,2),
    #dense
    layers.Flatten(),
    layers.Dense(10,activation="sigmoid"),
    
    keras.layers.Dense(10,activation="softmax")
])
```

```python
cnn.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
cnn.fit(x_train,y_train,epochs=10)
```
